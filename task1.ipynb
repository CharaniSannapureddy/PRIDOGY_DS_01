```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your dataset is stored in a DataFrame named df

# Group by country and sum the sales amount
sales_by_country = df.groupby('Country')['Amount'].sum().reset_index()

# Select the top 10 countries by total sales amount
top_10_countries = sales_by_country.nlargest(10, 'Amount')

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(top_10_countries['Country'], top_10_countries['Amount'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Total Sales Amount ($)')
plt.title('Top 10 Countries by Total Sales Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt

# Plotting
plt.figure(figsize=(10, 6))
plt.hist(df['Amount'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Sales Amount ($)')
plt.ylabel('Frequency')
plt.title('Distribution of Sales Amount')
plt.grid(True)
plt.tight_layout()
plt.show()
```
